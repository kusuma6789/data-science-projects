```python
import pandas as pd

data = {
    "Name": ["Asha", "Ravi", "Priya", "Kumar", "Neha", "Suman"],
    "Math": [78, 45, 92, 58, 88, None],
    "Science": [85, 56, 88, None, 92, 65],
    "English": [82, 60, 95, 70, 90, 60]
}

df = pd.DataFrame(data)
df.to_csv("student_scores.csv", index=False)
df
df = pd.read_csv("student_scores.csv")
print("Original Data:")
print(df)
for col in ['Math','Science','English']:
    df[col] = pd.to_numeric(df[col], errors='coerce')
    mean_val = df[col].mean()
    df[col].fillna(mean_val, inplace=True)

print("\nCleaned Data:")
print(df)
df['Average'] = df[['Math','Science','English']].mean(axis=1)
df
top = df.loc[df['Average'].idxmax()]
print(f"Top performer: {top['Name']} with average score {top['Average']:.2f}")
import matplotlib.pyplot as plt

plt.bar(df['Name'], df['Average'])
plt.title("Average Scores per Student")
plt.xlabel("Student")
plt.ylabel("Average Marks")
plt.show()
df.to_csv("student_scores_cleaned.csv", index=False)
```
